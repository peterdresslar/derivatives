```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")

```

This notebook is an exploration of derivative functions! According to [wikipedia](https://en.wikipedia.org/wiki/Derivative) a derivative is: a fundamental tool that quantifies the sensitivity to change of a function's output with respect to its input. Thatʻs actually a succinct and helpful definition! Letʻs take a look at some basic functions and the way derivatives "describe" function sensitivity.

Letʻs start with three very basic functions:

1. The constant function:
   $$f(x) = c$$
   
0. An arithmetic function:
   $$f(x) = x+x$$
   (which is the same as...)
   $$f(x) = 2x$$
   
2. The power function:
   $$f(x) = x^a$$

These are fundamental building blocks of calculus. Now, before we even take a look at the derivatives, letʻs think about what these functions look like on number lines.

First, letʻs look at our constant function:

This is:
   $$f(x) = c$$

In this case, we will just set c to equal 2. But it could be any number you like.

```python
plt.figure(figsize=(8, 5))
x = np.linspace(-5, 5, 100)

# Constant function f(x) = c (using c = 2)
c = 2
y = np.full_like(x, c)
plt.plot(x, y, 'b-', label=f'f(x) = {c}')
plt.grid(True)
plt.title('Constant Function')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.legend()
plt.show()
```

No surprises there, right? If the output of the function is always the same no matter what x is, of course our graph is just a straight, horizontal line.

Letʻs move on to our "geometric function"... weʻre just adding x to x here:

```python
plt.figure(figsize=(8, 5))
x = np.linspace(-5, 5, 100)

# Linear function f(x) = 2x
y = 2 * x
plt.plot(x, y, 'r-', label='f(x) = 2x')
plt.grid(True)
plt.title('Linear Function')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.legend()
plt.show()
```

And then letʻs take a look at our exponential function,    
$$f(x) = x^a$$

```python
plt.figure(figsize=(8, 5))
x = np.linspace(-5, 5, 100)

# Power function f(x) = x^a
a = 2
y = x**a
plt.plot(x, y, 'g-', label=f'f(x) = x^{a}')
plt.grid(True)
plt.title('Power Function')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.legend()
plt.show()
```

So, three functions: 

1. a straight horizontal line
2. a straight line going somewhere other than horizontal, and
3. a "power" curve, a standard-looking parabola curve

Before we *compute* the derivative to each of these, letʻs just talk about the answers. They are:

1. 0
2. 2
3. $$a\,x^{a-1}$$

What? This is so confusing! How can the answer be numbers for two of the functions/graphs, and something totally different for another? Well, hang in there! Letʻs take a deeper look at how we got our answers.

Again, a derivative tells us how "sensitive" a function is to the inputs. In other words, if we start plugging in lots of numbers, is the direction of the graph, say going from left to right, suddenly going to start having a bunch of rises and falls, and will that change with more inputs that take the graph further toward the right. So the ways we calculate what is actually happening with a graph have a lot to do with "what direction is it going next?"

For instance, in our first function, x=2. That means that no matter what number we put into x, we know that our output (known as f(x), which you are welcome to just think of as "y" since we are graphing) will always be 2. So, if x is 1, y is 2. If x is 3, y is 2. If x is a jillion, y is 2. Since a derivative describes what direction the function is going next... well, for x=2, the function isnʻt going *anywhere*. And so the derivative is just zero.

We have more than one way to write a derivative. Let's start with the simpler notation (called Lagrange notation):
$$f'(x)$$

This just means "the derivative of f". Simple! 

Later, you'll also see this notation (called Leibniz notation):
$$\frac{d}{dx}[f(x)]$$

They mean exactly the same thing! Mathematicians just developed different ways of writing it over the centuries. Think of it like how "2+2" and "sum(2,2)" mean the same thing - just different ways to write it.

For our constant function $f(x) = 2$, two ways to look at the derivative of that function are:
$$f'(x) = 0$$
or
$$\frac{d}{dx}[c] = 0$$

Again, the derivative itself is *zero*.





So basically, sometimes when we run a derivative function on another function, we get just a number. But, if the other function is *complicated* enough--if it does different stuff at different places, we need another function: a *derivative* function that basically tells us... "well, when we look over here, the function is acting like this. But when we look over here, the function is acting like this. And so you basically turn that discussion into a function on its own.... itʻs like "the rules" for how the base function "works."
